```python
%matplotlib inline
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
nfl_ranks = pd.read_csv('ranks3.csv')
```

```python
nfl_ranks.head()
```

```python
NFL = nfl_ranks[nfl_ranks['League']=='NFL']
```

```python
NFL
```

```python
NFL['pollution_rank_NFL'] = NFL['pollution_rank'].rank(method='first', ascending = True)
```

```python
NFL
```

```python
NFL['touchdowns_rank_NFL'] = NFL['touchdowns_rank'].rank(method='first', ascending = True)
```

```python
NFL[['touchdowns_rank','touchdowns_rank_NFL']]
```

```python
NFL['ticket_rank_NFL'] = NFL['ticket_rank'].rank(method='first', ascending = True)
```

```python
NFL['precipitation_rank_NFL'] = NFL['precipitation_rank'].rank(method='first', ascending = True)
```

```python
NFL['mcd_rank_NFL'] = NFL['mcd_rank'].rank(method='first', ascending = True)
```

```python
NFL.head()
```

```python
NFL['beer_rank_NFL'] = NFL['beer_rank'].rank(method='first', ascending = True)
```

```python
NFL
```

```python
NFL['sum_of_ranks_NFL'] = NFL['touchdowns_rank_NFL'] + NFL['ticket_rank_NFL'] + NFL['precipitation_rank_NFL'] + NFL['beer_rank_NFL'] + NFL['mcd_rank_NFL'] +NFL['pollution_rank_NFL']
```

```python
NFL['total_rank_NFL'] = NFL['sum_of_ranks_NFL'].rank(method='first', ascending = True)
```

```python
NFL
```

```python
NFL.drop(['touchdowns_rank','ticket_rank','precipitation_rank','beer_rank','mcd_rank','pollution_rank','sum_of_ranks','total_rank'], axis = 1, inplace=True)
```

```python
NFL.columns
```

```python
NFL.set_index('Unnamed: 0', inplace=True)
```
