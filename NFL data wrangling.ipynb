```python
nfl_ranks = nfl_data[['Team','City','touchdowns_rank','ticket_rank','precipitation_rank','beer_rank','mcd_rank','pollution_rank']]
```

```python
nfl_ranks.head()
```

```python
nfl_ranks['sum_of_ranks'] = nfl_ranks['touchdowns_rank'] + nfl_ranks['ticket_rank'] + nfl_ranks['precipitation_rank'] + nfl_ranks['beer_rank'] + nfl_ranks['mcd_rank'] + nfl_ranks['pollution_rank']
```

```python
nfl_ranks.sample(10)
```

```python
nfl_ranks['total_rank'] = nfl_ranks['sum_of_ranks'].rank(method='first', ascending = True)
```
